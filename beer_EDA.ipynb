```python
#Also drop Gose, Cider and Barleywine

filtered_ids = list(beer_df[(beer_df.other == 1) | (beer_df.cider == 1)| (beer_df.gose == 1) |(beer_df.barleywine == 1)].beerid)
```

```python
#flip signs of beer filter above

beer_mask = beer_df.beerid.isin(filtered_ids) == False
```

```python
final_beer_df = beer_df[beer_mask]
```

```python
final_beer_df
```

```python
final_beer_counts = final_beer_df.iloc[:,21:].sum()
final_beer_counts
```

```python
final_beer_counts.sum()
```

```python
pd.Series.apply(final_beer_counts, lambda x: x / final_beer_counts.sum())
```

```python
final_beer_df = final_beer_df.drop(columns = ['cider', 'gose', 'barleywine', 'other'])
```
